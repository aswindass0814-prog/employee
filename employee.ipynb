```python
# outlier handling


Q1=df['Monthly Hours'].quantile(0.25)
Q3=df['Monthly Hours'].quantile(0.75)

IQR=Q3-Q1

lower= Q1-1.5*IQR
upper=Q3+1.5*IQR

(lower,upper)

me1=df['Monthly Hours'].mean()

for i in df.index:
    if df.loc[i,'Monthly Hours']>upper or df.loc[i,'Monthly Hours']<lower:
        df.loc[i,'Monthly Hours']=me1



#df=df[(df['backrest_angle_deg']>=lower)&(df['backrest_angle_deg']<=upper)]

sns.boxplot(df['Monthly Hours'])
```

```python
#encoding

from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()

df['Gender']=lab.fit_transform(df['Gender'])
df['Company']=lab.fit_transform(df['Company'])
df[ 'Department']=lab.fit_transform(df[ 'Department'])
```

```python
df.drop(['Name','Emp_id'],axis=1,inplace=True)
```

```python
df.columns
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
x_train
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
# decison tree

from sklearn.tree import DecisionTreeClassifier
tr=DecisionTreeClassifier()
tr.fit(x_train,y_train)
y_pred2=tr.predict(x_test)
```

```python
ac2=accuracy_score(y_pred2,y_test)
ac2
```

```python
# random forest

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(class_weight='balanced')
rf.fit(x_train,y_train)
y_pred3=rf.predict(x_test)
y_pred3
```

```python
ac3=accuracy_score(y_test,y_pred3)
ac3
```

```python
#make this model more accurate



from sklearn.svm import SVC
svm = SVC()

svm.fit(x_train,y_train)
y_pred4=svm.predict(x_test)
```

```python
ac4=accuracy_score(y_test,y_pred4)
ac4
```

```python
model=[y_pred2,y_pred3]
ac=[ac2,ac3,ac4]

plt.bar(['Decision Tree','Random Forest','SVM'],[ac2,ac3,ac4],color=['r'])
plt.xlabel('Models')
plt.ylabel('Score')
plt.title('Model Score')
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred5=reg.predict(x_test)  
y_pred5
```

```python
from sklearn.metrics import r2_score
r1=r2_score(y_test,y_pred5)
r1
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred5)
mse
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape=mean_absolute_percentage_error(y_test,y_pred5)
mape
```
